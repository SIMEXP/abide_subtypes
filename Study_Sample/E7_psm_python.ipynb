```python
# Imports
import math
import numpy as np
import scipy as scp
import sklearn as skl
from scipy import stats as st
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Paths
pheno_path = '/data1/abide/Pheno//full_merged_pheno.csv' # ABIDE with ADOS_sev
```

```python
# Get the pheno
pheno = pd.read_csv(pheno_path)
# Kill whitespace
pheno.columns = [x.strip(' ') for x in pheno.columns]
# select the columns that need to go into the PSM
cols = ['SITE_ID', 'SUB_ID', 'AGE_AT_SCAN', 
        'FD', 'FD_scrubbed', 'EYE_STATUS_AT_SCAN',
        'FIQ', 'VIQ', 'PIQ', 'DX_GROUP']
use_pheno = pheno[cols]
use_pheno['DX_GROUP'].replace(to_replace=[2,1], value=[1,0], inplace=True)
```

```python
nyu_pheno = use_pheno[use_pheno.SITE_ID=='NYU']
```

```python
nyu_pheno.head()
```

```python
noise = ['AGE_AT_SCAN', 'FD_scrubbed', 'FIQ']
```

```python
# Compute the propensity score
# Fit the log model
propensity = skl.linear_model.LogisticRegression().fit(nyu_pheno[noise], nyu_pheno.DX_GROUP)
# Predict the propensity score
pscore = propensity.predict_proba(nyu_pheno[noise])
```
