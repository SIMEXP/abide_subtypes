```python
import os
import numpy as np
import pandas as pd
import nibabel as nib
import nilearn as nil
from scipy import io as sio
from matplotlib import gridspec
from nilearn import plotting as nlp
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
in_path = '/home/surchs/VM_Transfer/subtypes/TRT'
```

```python
res_img = nib.load(os.path.join(in_path, 'base_1_net_1_sbt_1.nii.gz'))

fig = plt.figure(figsize=(30,25))
gs = gridspec.GridSpec(10, 2, hspace=0)
# Show the average map
for sbt_id, fig_pos in enumerate(np.array([0, 2, 4, 6, 8])):
    a_img = nib.load(os.path.join(in_path, 'base_1_net_1_sbt_{}.nii.gz'.format(sbt_id+1)))
    ax_a = fig.add_subplot(gs[fig_pos,0])
    ax_b = fig.add_subplot(gs[fig_pos+1,0])
    tmp = nlp.plot_stat_map(a_img, display_mode='x', alpha=0.9,
                            cut_coords=np.arange(60,0,-10), black_bg=True, axes=ax_a,
                            title='Subtype {}'.format(sbt_id+1))
    tmp = nlp.plot_stat_map(a_img, display_mode='x', alpha=0.9,
                            cut_coords=np.arange(-60,0,10), black_bg=True, axes=ax_b, colorbar=False)
    
    a_img = nib.load(os.path.join(in_path, 'base_2_net_1_sbt_{}.nii.gz'.format(sbt_id+1)))
    ax_a = fig.add_subplot(gs[fig_pos,1])
    ax_b = fig.add_subplot(gs[fig_pos+1,1])
    tmp = nlp.plot_stat_map(a_img, display_mode='x', alpha=0.9,
                            cut_coords=np.arange(60,0,-10), black_bg=True, axes=ax_a,
                            title='Subtype {}'.format(sbt_id+1))
    tmp = nlp.plot_stat_map(a_img, display_mode='x', alpha=0.9,
                            cut_coords=np.arange(-60,0,10), black_bg=True, axes=ax_b, colorbar=False)
```
